```python
docs = [
  {

      "id" : 1,
      "body" : "The star in our solar system provides heat and light."

  },
  {

      "id" : 2,
      "body" : "That Hollywood star walked the red carpet for the movie premiere."

  },
  {

      "id" : 3,
      "body" : "Astronomers observe distant stars and galaxies using telescopes."

  }
]

```

```python
from sys import version_info
import argparse
import json
import matplotlib.pyplot as plt
import os
import sys
sys.path.append(os.path.abspath(r'C:\Users\srksr\NLP\NLP_project\NLP_ASSIGNMENT2'))
from sentenceSegmentation import SentenceSegmentation
from tokenization import Tokenization
from inflectionReduction import InflectionReduction
from stopwordRemoval import StopwordRemoval
from informationRetrieval import InformationRetrieval
from evaluation import Evaluation
```

```python
tokenizer = Tokenization()
sentSeg = SentenceSegmentation()
inflRed = InflectionReduction()
stopwordRem = StopwordRemoval()
ir = InformationRetrieval()
eval = Evaluation()
```

```python
docids = [doc["id"] for doc in docs]
segmentedDocs = []
for doc in docs:
    segmentedDoc = sentSeg.punkt(doc["body"])
    segmentedDocs.append(segmentedDoc)


tokenizedDocs = []
for doc in segmentedDocs:
    tokenizedDoc = tokenizer.spacyTokenizer(doc)
    tokenizedDocs.append(tokenizedDoc)


reducedDocs = []
for doc in tokenizedDocs:
    reducedDoc = inflRed.reduce(doc)
    reducedDocs.append(reducedDoc)

stop_words=["the", "in", "our", "and", "that", "for"]
stopwordRemovedDocs = []
for doc in reducedDocs:
    stopwordRemovedDoc = stopwordRem.fromList(doc, stop_words)
    stopwordRemovedDocs.append(stopwordRemovedDoc)

```

```python
stopwordRemovedDocs
```

```python
ir.buildIndex(stopwordRemovedDocs,docids)
ir.index
```

```python
import numpy as np
tf_matrix = np.zeros_like(ir.tfidf).T
for i in range(len(ir.words)):
    for j in range(ir.total_docs):
        if ir.words[i] in ir.index and docids[j] in ir.index[ir.words[i]]:
            tf_matrix[i][j] = ir.index[ir.words[i]][docids[j]]
```

```python
query = ["slip-flow heat transfer in internal channels"]
processedQuery = stopwordRem.fromList(inflRed.reduce(tokenizer.spacyTokenizer(query)), stop_words)
print(processedQuery)


doc_list = {}
for words in processedQuery:
    for word in words:
        if word in list(ir.index.keys()):
            doc_list[word] = list(ir.index[word].keys())
            doc_list[word].remove('idf')
print(doc_list)
```

```python
querytfidf = np.zeros((1, len(ir.words)))
for i in range(len(ir.words)):
    if ir.words[i] in processedQuery[0]:
        querytfidf[0][i] += ir.index[ir.words[i]]['idf']
```

```python
querytfidf
```

```python
normalisedQuery = querytfidf / np.linalg.norm(querytfidf)
normalisedtfidf_docs = ir.normalized_tfidf
similarity_scores = np.dot(normalisedtfidf_docs, normalisedQuery.T).flatten()
similarity_scores
```
